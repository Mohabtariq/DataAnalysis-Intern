# 2 - Data Cleaning
 

```python
print(df.duplicated().sum())
print(df.isnull().any())
print(df.isnull().sum())
```

## 2-1 check Data Types

```python
cols = ['BDMG_0', 'BDMG_1', 'BDMG_2', 'BDMG_3', 'HDMG_0', 'HDMG_1', 'HDMG_2', 'HDMG_3']
df[cols] = df[cols].apply(pd.to_numeric, errors='coerce')
```

## 2-2 Clean Nulls 

```python

df['BDMG_0']=df['BDMG_0'].fillna(df['BDMG_0'].mean())
df['BDMG_1']=df['BDMG_1'].fillna(df['BDMG_1'].mean())
df['BDMG_2']=df['BDMG_2'].fillna(df['BDMG_2'].mean())
df['BDMG_3']=df['BDMG_3'].fillna(df['BDMG_3'].mean())
df['HDMG_0']=df['HDMG_0'].fillna(df['HDMG_0'].mean())
df['HDMG_1']=df['HDMG_1'].fillna(df['HDMG_1'].mean())
df['HDMG_2']=df['HDMG_2'].fillna(df['HDMG_2'].mean())
df['HDMG_3']=df['HDMG_3'].fillna(df['HDMG_3'].mean())
```

## Which weapon type has the highest bullet efficiency 

```python
df['Bullet_Efficiency'] = df['Damage'] / df['Magazine Capacity']      #Bullet Efficiency Formula = damage/magazine
avg_eff=df.groupby('Weapon Type')['Bullet_Efficiency'].mean().reset_index().sort_values(by='Bullet_Efficiency',ascending=False)
avg_eff=avg_eff[avg_eff['Weapon Type']!='Melee']
print(avg_eff)
sns.barplot(data=avg_eff, y='Weapon Type', x='Bullet_Efficiency')
plt.title("Bullet Efficiency by Weapon Type (Excluding Melee)")
plt.xlabel("Efficiency")
plt.ylabel("Weapon Type")
plt.show()
```

## Correlation 
- Do weapons with higher bullet capacities have better performance ?
- Is there a significant correlation between bullet type and any other weapon characteristic ?
- How does the rate of fire of a weapon relate to its damage per second ?

```python
num_cols=['Damage','Magazine Capacity','Range','Bullet Speed','Rate of Fire','Damage Per Second','Bullet Type','Shots to Kill (Chest)','Shots to Kill (Head)']
df_numeric = df[num_cols]
correlation_matrix = df_numeric.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, square=True, fmt=".2f",
            linewidths=0.5, linecolor='gray', cbar_kws={"shrink": 0.75})
plt.title("Correlation Matrix for Weapon Characteristics")
plt.show()
```

Range and Bullet Speed (0.84) 
- Negative Corr \
Rate of Fire and Damage Per Second (-0.61) \
Magazine Capacity and Damage Per Second (-0.38)

## Which specific weapons are associated with each bullet type?

```python
unique_bullet_types = df['Bullet Type'].unique()
unique_bullet_types = unique_bullet_types[pd.notna(unique_bullet_types)]


for bullet in unique_bullet_types:
    weapon_names = df.loc[df["Bullet Type"] == bullet, "Weapon Name"]
    plt.figure(figsize=(8, 6))
    plt.pie([1]*len(weapon_names), labels=weapon_names, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("Set3", len(weapon_names)))
    plt.title(f'Weapons using {bullet} Bullet Type')
    plt.show()
```

## Average Damage Reduction considering (Vest,Helmet) vs Weapon type

```python
# How much more protection does chest armor provide compared to helmet armor for each weapon?
# each level compared to level 0 that mean no helmet or chest 
df['BDMG_Diff_1'] = df['BDMG_1'] - df['BDMG_0']
df['BDMG_Diff_2'] = df['BDMG_2'] - df['BDMG_0']
df['BDMG_Diff_3'] = df['BDMG_3'] - df['BDMG_0']
df['HDMG_Diff_1'] = df['HDMG_1'] - df['HDMG_0']
df['HDMG_Diff_2'] = df['HDMG_2'] - df['HDMG_0']
df['HDMG_Diff_3'] = df['HDMG_3'] - df['HDMG_0']
diff_columns = ['BDMG_Diff_1', 'BDMG_Diff_2', 'BDMG_Diff_3', 
                   'HDMG_Diff_1', 'HDMG_Diff_2', 'HDMG_Diff_3']

df[diff_columns] = df[diff_columns].apply(pd.to_numeric, errors='coerce')
damage_diff = df.groupby('Weapon Type')[diff_columns].mean().reset_index().sort_values(by=diff_columns,ascending=True)
damage_diff
damage_diff['BDMG_Mean']=(damage_diff['BDMG_Diff_1']+damage_diff['BDMG_Diff_2']+damage_diff['BDMG_Diff_3']//3)
damage_diff['HDMG_Mean']=(damage_diff['HDMG_Diff_1']+damage_diff['HDMG_Diff_2']+damage_diff['HDMG_Diff_3']//3)
print(damage_diff['BDMG_Mean'].describe())
print(damage_diff['HDMG_Mean'].describe())
fig, axs = plt.subplots(2, 1, figsize=(8, 10))
sns.barplot(data=damage_diff,x='BDMG_Mean',y='Weapon Type',palette='coolwarm',ax=axs[0])
axs[0].set_title('Body Damge Reduction by Weapon Type')
axs[0].set_xlabel('Body Damge Reduction')
axs[0].set_ylabel('Weapon Type')


sns.barplot(data=damage_diff,x='HDMG_Mean',y='Weapon Type',palette='coolwarm',ax=axs[1])
axs[1].set_title('Head Damge Reduction by Weapon Type')
axs[1].set_xlabel('Body Damge Reduction')
axs[1].set_ylabel('Weapon Type')

plt.show()
```

- ## By considering all the factors Shotgun is power full in short range close fights around 25-30
- ## AR type guns are suitable for Short to average range fights
- ## AWM is most powerfull gun for Long range fights
- ## most reducted weapon is the shotgun follwed by sniper
